```python
import random as rn
import numpy as np
import pandas as pd

cov_matrix = np.array([[1.61, -0.14, -1.03],
                     [-0.14, 2.85, 0.40],
                     [-1.03, -0.64, 1.52]])

X = np.array(['X5', 'X6', 'X8'])
df = pd.DataFrame(cov_matrix, X, X)
mean_values = np.array([0.5, 2.3, 4.35])
df
```

```python
np.random.seed(42)
sample = np.random.multivariate_normal(mean_values, cov_matrix, 100)#выборка многомерной нормальной случайной величины
sample_df = pd.DataFrame(sample, range(100), X)
sample_df
```

```python

sample_cov_matrix = np.cov(sample.T, bias=True)#Матрица ковариаций для выборочных зачений (смещенная оценка)
print('         Матрица ковариаций:')
df = pd.DataFrame(sample_cov_matrix, X, X)
df
```

```python
print('Проверка гипотезы о равенстве матрицы ковариаций заданной матрице')
print('Отношение правдоподобия для данной гипотезы:')
n = 100 #объем выборки
k = 3
l = sample_cov_matrix @ np.linalg.inv(cov_matrix)
lambda_1 = np.linalg.det(l)**(n/2) * np.exp(1/2*k*n - 1/2*n*np.trace(l))

print(f'lambda =', lambda_1)
etta = - 2*np.log(lambda_1) #статистика
print(f'etta =', etta)

```

```python
from scipy.stats.distributions import chi2
#проверка гипотезы на уровне значимости 0.06
nu = 1/2*k*(k+1)
chi_2 = chi2.ppf(0.94, df=nu) #критическое значение
print('Критическое значение:', chi_2)
if etta<chi_2:
  print('Гипотеза H0 принимается')
else:
  print('Гипотеза H0 отвергается')
```

```python
print('Проверка гипотезы о равенстве матриц ковариаций и векторов средних')
sample_mean = sample.mean(axis = 0) #Выборочное среднее по столбцам
print('Выборочные средние [X5, X6, X8]:', sample_mean)
mean_diff = sample_mean - mean_values
lambda_2 = np.linalg.det(l)**(n/2)* np.exp(1/2*n*(k - np.trace(l) - mean_diff.T @ np.linalg.inv(sample_cov_matrix) @ mean_diff))
print(f'lambda =', lambda_2)
etta_2 = - 2*np.log(lambda_2) #статистика
print(f'etta =', etta_2)
```

```python
#проверка гипотезы на уровне значимости 0.06
nu = 1/2*k*(k+1)+ k
chi_2 = chi2.ppf(0.94, df=nu) #критическое значение

print(chi_2)
if chi_2 > etta_2 :
  print('Гипотеза H0 принимается')
else:
  print('Гипотеза H0 отвергается')
print('Вывод: матрицы ковариаций и векторы средних равны')
```
